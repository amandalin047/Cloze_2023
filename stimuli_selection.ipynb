```python
def REINDEX(array):
    '''Sorts sentences by a given index in asencing order'''
    array_ascend = np.sort(array)
    unique, idx = [], []
    for i, x in enumerate(array_ascend):
        if x not in unique:
            idx.append(np.where(array==x)[0].tolist())
            unique.append(x)
    return sum(idx, [])

def main(df1, df2, plaus_thresh=5, val_thresh=2, condition=None):
    df = pd.concat((df1, df2));
    
    df = df[df['Plaus. mean']>=plaus_thresh]
    if condition == 'Emo':
        df = df[(df['Val. mean']>=5+val_thresh) | (df['Val. mean']<=5-val_thresh)].reset_index(drop=True)
    elif condition == 'Neu':
        df = df[(df['Val. mean']<=5+val_thresh) & (df['Val. mean']>=5-val_thresh)].reset_index(drop=True)
    df = df.reindex(REINDEX(df['Index'])).reset_index(drop=True)
    
    idx = df['Index'].tolist()
    for i, x in enumerate(idx):
        if idx.count(x) == 1:
            df = df.drop(index=i)
    
    sc, wc = df[(df['Index']%4==0) | (df['Index']%4==1)], df[(df['Index']%4==2) | (df['Index']%4==3)]
    print('number of '+condition+'_SC frames =', int(len(df[(df['Index']%4==0) | (df['Index']%4==1)])/2))
    display(df[(df['Index']%4==0) | (df['Index']%4==1)])
    print('\n\nnumber of '+condition+'_WC frames =', int(len(df[(df['Index']%4==2) | (df['Index']%4==3)])/2))
    display(df[(df['Index']%4==2) | (df['Index']%4==3)])
    
    return sc, wc
```

```python
os.chdir('/home/amandalin047/july_ratings/results/1st_round')
df1 = pd.read_excel('1st_all_means.xlsx')
os.chdir('/home/amandalin047/july_ratings/results/2nd_round')
df2 = pd.read_excel('2nd_all_means.xlsx')

emo_sc, emo_wc = main(df1, df2, plaus_thresh=4.5, val_thresh=1.7, condition='Emo');
neu_sc, neu_wc = main(df1, df2, plaus_thresh=4.5, val_thresh=1, condition='Neu');
```

```python
os.chdir('/home/amandalin047/july_ratings')
to_export, outfile_names = [emo_sc, emo_wc, neu_sc, neu_wc], ['emo_sc.xlsx', 'emo_wc.xlsx', 'neu_sc.xlsx', 'neu_wc.xlsx']
for i, x in enumerate(to_export):
    x.to_excel(outfile_names[i], index=False)
```

```python
df = pd.DataFrame(None, columns=emo_sc.columns)
for i in to_export:
    df = pd.concat((df, i))
display(df)

print('CORRELATION')
display(df[['Exp. mean', 'Plaus. mean']].corr(numeric_only=False))

import matplotlib.pyplot as plt

figure, axis = plt.subplots(1,2, figsize=(13,5))
axis[0].scatter(df['Cloze'], df['Exp. mean'])
axis[0].set_xlabel('Cloze')
axis[0].set_ylabel('Expectancy')
axis[0].set_title('Cloze vs Expectancy', fontsize=16)
axis[1].scatter(df['Plaus. mean'], df['Exp. mean'])
axis[1].set_title('Plausibility vs Expectancy', fontsize=16)
axis[1].set_xlabel('Plausibility')
axis[1].set_ylabel('Expectancy')
plt.show()
```
