```python
import pandas as pd
import os
import numpy as np
```

```python
def REINDEX(array):
    '''Sorts sentences by a given index in asencing order'''
    array_ascend = np.sort(array)
    unique, idx = [], []
    for i, x in enumerate(array_ascend):
        if x not in unique:
            idx.append(np.where(array==x)[0].tolist())
            unique.append(x)
    return sum(idx, [])
```

```python
def main(df1, df2, exp_thresh=4.5, plaus_thresh=5, val_thresh=2, condition=None):
    df = pd.concat((df1, df2)).reset_index(drop=True);
    df = df.reindex(REINDEX(df['Cloze'])).reset_index(drop=True);
    
    num_zero_cloze = len([i for i in df['Cloze'].tolist() if i==0])
    new_df = pd.DataFrame(None, columns=df.columns)
    for i in range(len(df.head(num_zero_cloze))):
        if df['Exp. mean'].iloc[i] <= exp_thresh: new_df.loc[len(new_df)] = df.iloc[i]     

    for i in range(len(df.head(num_zero_cloze)),len(df)):
        if df['Index'].iloc[i] in new_df['Index'].tolist(): new_df.loc[len(new_df)] = df.iloc[i]
    
    new_df = new_df.reindex(REINDEX(new_df['Index'])).reset_index(drop=True)
    
    new_df = new_df[new_df['Plaus. mean']>=plaus_thresh]
    if condition == 'Emo':
        new_df = new_df[(new_df['Val. mean']>=5+val_thresh) | (new_df['Val. mean']<=5-val_thresh)].reset_index(drop=True)
    elif condition == 'Neu':
        new_df = new_df[(new_df['Val. mean']<=5+val_thresh) & (new_df['Val. mean']>=5-val_thresh)].reset_index(drop=True)
    new_df = new_df.reindex(REINDEX(new_df['Index'])).reset_index(drop=True)
    
    idx = new_df['Index'].tolist()
    for i, x in enumerate(idx):
        if idx.count(x) == 1:
            new_df = new_df.drop(index=i)
    
    sc, wc = new_df[(new_df['Index']%4==0) | (new_df['Index']%4==1)], new_df[(new_df['Index']%4==2) | (new_df['Index']%4==3)]
    print('number of '+condition+'_SC frames =', int(len(new_df[(new_df['Index']%4==0) | (new_df['Index']%4==1)])/2))
    display(new_df[(new_df['Index']%4==0) | (new_df['Index']%4==1)])
    print('\n')
    print('number of '+condition+'_WC frames =', int(len(new_df[(new_df['Index']%4==2) | (new_df['Index']%4==3)])/2))
    display(new_df[(new_df['Index']%4==2) | (new_df['Index']%4==3)])
    print('\n')
    
    return sc, wc
```

```python
os.chdir('/home/amandalin047/july_ratings/results/1st_round')
df1 = pd.read_excel('1st_all_means.xlsx')
os.chdir('/home/amandalin047/july_ratings/results/2nd_round')
df2 = pd.read_excel('2nd_all_means.xlsx')

emo_sc, emo_wc = main(df1, df2, exp_thresh=5.75, plaus_thresh=3.7, val_thresh=1.6, condition='Emo');
neu_sc, neu_wc = main(df1, df2, exp_thresh=5.75, plaus_thresh=3.7, val_thresh=0.9, condition='Neu');
```

```python
os.chdir('/home/amandalin047/july_ratings')
to_export, outfile_names = [emo_sc, emo_wc, neu_sc, neu_wc], ['emo_sc.xlsx', 'emo_wc.xlsx', 'neu_sc.xlsx', 'neu_wc.xlsx']
for i, x in enumerate(to_export):
    x.to_excel(outfile_names[i], index=False)
```

```python
import pingouin as pg

all_exp = df.iloc[[i for i in range(len(df)) if df['Cloze'].iloc[i]!=0]]
all_unexp = df.iloc[[i for i in range(len(df)) if df['Cloze'].iloc[i]==0]]

table = pd.DataFrame({'Exp. Sentences': [all_exp['Plaus. mean'].mean(), all_exp['Exp. mean'].mean()],
                      'Unexp. Sentences': [all_unexp['Plaus. mean'].mean(), all_unexp['Exp. mean'].mean()]},
                       index=['Plausibility', 'Expectancy'])
display(table)

print('\n\nT-testing whether the rated plausibility of expected vs unecpected sentences significantly differed')
display(pg.ttest(all_exp['Plaus. mean'].tolist(), all_unexp['Plaus. mean'].tolist()))
print('\nT-testing whether the rated expectancy of expected vs unecpected sentences significantly differed')
display(pg.ttest(all_exp['Exp. mean'].tolist(), all_unexp['Exp. mean'].tolist()))

print('\nAll expected plausibility mean = ', all_exp['Plaus. mean'].mean())
print('All unexpecte', all_unexp['Plaus. mean'].mean())
```
